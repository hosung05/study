```python
def age_range(x):
    if int(x) < 10:
        return '유아'
    elif int(x) >= 70:
        return '노인'
    else:
        return f'{int(x)//10}0대'
# age_range(71)
# np.select()
df2['Age_Range'] = df2['Age'].apply(age_range)
df2.info()
df2.groupby('Age_Range').sum()['Survived']
```
