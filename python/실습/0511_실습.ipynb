```python
age = 20
gender = "F"
if 20 <= age < 40:
    if gender == "M":
        print("hobby")
    elif gender == "F" and age == "unmarrige":
        print("trip")
    else:
        print("info")
elif 40 < age <= 60:
    print('education')
elif age > 60:
    print("health")
else:
    print("해당사항이 없습니다.")
```

```python
score1 = input("첫번째 점수를 입력해주세요")
score2 = input("두번째 점수를 입력해주세요")
score3 = input("세번째 점수를 입력해주세요")

score1 = int(score1)
score2 = int(score2)
score3 = int(score3)

if not (0 <= score1 <= 100) or not ( 0 <= score2 <= 100) or not (0 <= score3 <= 100):
    print("잘못된 점수 입니다")
elif score1 > 65 and score2 > 65 and score3 > 65:
    print('합격')
else:
    print("불합격")
```
