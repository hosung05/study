```python
twitter = Twitter()
komoran = Komoran()
kkma = Kkma()
```

```python
burgerking['store_name'].unique()
burgerking_reviews_dict = {}
for store in burgerking['store_name'].unique():
    temp_list = []
    for review in burgerking[burgerking['store_name'] == store]['reviews'] :
        temp_list.append(twitter.nouns(review))
    burgerking_reviews_dict[store] = temp_list
burgerking_reviews_dict
```

```python
lotteria['store_name'].unique()
lotteria_reviews_dict = {}
for store in lotteria['store_name'].unique():
    temp_list = []
    for review in lotteria[lotteria['store_name'] == store]['reviews'] :
        temp_list.append(twitter.nouns(review))
    lotteria_reviews_dict[store] = temp_list
lotteria_reviews_dict
```
